```python
# cost function

# samples = data.values
# cost function is J(w, b)
def cost_function(w, b, samples):
  total_cost = 0
  m = len(samples)

  for i in range(m):
    x = samples[i, 0]
    y = samples[i, 1]

    total_cost += (w * x + b - y)**2

  total_cost /= (2 * m)
  return total_cost
```

```python
# gradient descent: updates values of parameters w, b and keeps iterating unless convergence (minimized cost function)

def gradient_descent(w, b, samples, L): # L is the learning rate
  w_grad = 0 # w_gradient is d(J)/d(w)
  b_grad = 0 # b_gradient is d(J)/d(b)
  m = len(samples)

  for i in range(m):
    x = samples[i, 0]
    y = samples[i, 1]

    w_grad += 1/m * (w * x + b - y) * x # (solved it mathematically)
    b_grad += 1/m * (w * x + b - y) # (solved it mathematically)


  w -= L * w_grad # gradient descent formula
  b -= L * b_grad
  return w, b
```

```python
# running the linear regression model

w = 0
b = 0
L = 0.0001
epochs = 1000
samples = data.values

for i in range(epochs):
  w, b = gradient_descent(w, b, samples, L)

  if i % 100 == 0:
    print(f'epoch {i}: w = {w}, b = {b}, cost = {cost_function(w, b, samples)}')

plt.scatter(data['Study Time (hours)'], data['Score'])
plt.plot(data['Study Time (hours)'], w * data['Study Time (hours)'] + b)
plt.show()
```
